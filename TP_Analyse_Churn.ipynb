# **Travaux pratiques : analyse du churn**

Dans ce TP, vous allez travailler sur un jeu de données clients issu d’une entreprise de services (télécom, banque ou abonnement).  
L’objectif est d’analyser les caractéristiques des clients et de comprendre les facteurs liés au **churn**, c’est-à-dire le fait qu’un client quitte le service.

Le jeu de données contient :

- `Names` : nom du client

- `Age` : âge du client

- `Total_Purchase` : montant total des achats réalisés par le client

- `Account_Manager` : indique si le client dispose d’un account manager (`1`) ou non (`0`)

- `Years` : ancienneté du client en années

- `Num_Sites` : nombre de sites visités par le client

- `Onboard_date` : date d’entrée du client

- `Location` : localisation ou adresse du client

- `Company` : nom de l’entreprise cliente

- `Churn` : variable cible indiquant si le client a quitté le service (`1`) ou non (`0`)

La variable d'intérêt est la variable **Churn**.
