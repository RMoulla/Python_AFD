## Les conditions

```python
# si a est supérieur à 1 afficher "a est supérieur à 1"

if a > 1:
   print("a est supérieur à 1")
```

```python
# Si a est supérieur à 3 afficher "à supérieur à 3"
# sinon afficher "a est inférieur ou égal à 3"

if a > 3:
    print("a est supérieur à 3")
else:
    print("a est inférieur ou égal à 3")
```

```python
# Si a est égal à 2 afficher "a est égal à 2"

if a == 2:
    print("a est égal à 2")
```

```python
# Condition complexe :
# Si a n'est pas égal à 1
# et si a est inférieur à 3
# afficher a est inférieur à 3

if a != 1 and a < 3:
    print("a est inférieur à 3")
```

```python
# Si a est différent de 1
# ou si a est inférieur 0
# afficher "Bonjour"

if a != 1 or a < 0:
    print("Bonjour")
```

```python
# Si a est égal à 1 afficher "a est égal à 1"
# sinon si a est inférieur à 1 afficher "a est inférieur à 1"
# sinon si a est supérieur à 1 afficher "a est supérieur à 1"

if a == 1:
  print("a est égal à 1")
elif a < 1:
  print("a est inférieur à 1")
else:
  print("a est supérieur à 1")
```

## Les listes

```python
# Définir une liste

ma_liste = [10, 20, 40, 20, 5]

print(ma_liste)
```

```python
# Accéder au premier élément de la liste

ma_liste[0]
```

```python
# Accéder au deuxième élément de la liste

ma_liste[1]
```

```python
# Accéder au dernier élément de la liste

ma_liste[-1]
```

```python
# Accéder aux plusieurs éléments en fonction de l'index

ma_liste[1:3]
```

```python
# Modifier un élément d'une liste

ma_liste[0] = 15
print(ma_liste)
```

```python
# Insérer un élément dans une liste

ma_liste.insert(2, 35)
print(ma_liste)
```

```python
# Ajouter un élément à la dernière position

ma_liste.append(20)
print(ma_liste)
```

```python
# Supprimer un élément d'une liste avec pop

ma_liste.pop(2)
print(ma_liste)
```

```python
# Supprimer l'élément 20

ma_liste.remove(20)
print(ma_liste)
```

```python
# Enlever tous les éléments dont la valeur est 20 de la liste
for i in ma_liste:
  if i == 20:
    ma_liste.remove(i)

print(ma_liste)
```

```python
# Ordonner une liste

ma_liste.sort()
print(ma_liste)
```

```python
# Inverser l'ordre des éléments

ma_liste.reverse()
print(ma_liste)
```

```python
# Ajouter une liste à une liste liste2 existante

liste2 = ["a", 40, "Bob"]
ma_liste.extend(liste2)
print(ma_liste)
```

```python
# Copier une liste

ma_copie = ma_liste
print(ma_copie)
```

```python
# Enlever le dernier élément de ma_copie

ma_copie.pop()
print(ma_copie)
```

```python
print(ma_liste)
```

```python
# Copier une liste avec copy

ma_copie = ma_liste.copy()
print(ma_copie)
```

```python
ma_copie.pop()
print(ma_copie)
print(ma_liste)
```

```python
# Calculer les carrés d'une liste et sauvegarder dans liste_carre

ma_liste = [1, 2, 3, 4, 5]

liste_carre = []

for i in ma_liste:
  liste_carre.append(i**2)

print(liste_carre)
```

```python
# Calculer les carrés avec une comprehension de liste

liste_carre = [i**2 for i in ma_liste]
print(liste_carre)
```

## Les dictionnaires

```python
# Définir un dictionnaire avec {}

mon_dict = {"Nom" : "Alice", "Age" : 38, "Ville" : "Lyon"}
print(mon_dict)
```

```python
# Accéder à la valeur de Age

mon_dict["Age"]
```

```python
# Modifier la valeur de l'âge

mon_dict["Age"] = 39
print(mon_dict)
```

```python
# Ajouter une clé profession

mon_dict["Profession"] = "Ingénieur"
print(mon_dict)
```

```python
# Récupérer les clés du dictionnaire

print(mon_dict.keys())
```

```python
# Sauvegarder les clés d'un dictionnaire dans une liste
mes_cles = list(mon_dict.keys())
print(mes_cles)
```

```python
mes_cles[0]
```

```python
# Sauvegarder les valeurs d'un dictionnaire dans une liste
mes_valeurs = list(mon_dict.values())
print(mes_valeurs)
```

```python
# Créer un dictionnaire à partir de deux listes

cles = ["Nom", "Age", "Ville"]
valeurs = ["Bob", 42, "Paris"]

bob_dict = dict(zip(cles, valeurs))
print(bob_dict)
```

## Les dataframes

```python
# Afficher les noms des colonnes

df.columns
```

```python
# Afficher les valeurs de la colonne Nom

df['Nom']
```

```python
# Extraire un sous-dataframe

sub_df = df[['Nom', 'Age']]

sub_df
```

```python
# Affichier les deux premières lignes

df.head(2)
```

```python
# Afficher les deux dernières lignes

df.tail(2)
```

```python
# Afficher la première ligne

df.iloc[0]
```

```python
# Afficher les lignes où l'âge est supérieur à 30 ans

df[df['Age'] > 30]
```

```python
# Afficher les lignes où Ville est Lyon

df[df['Ville'] == "Lyon"]
```
